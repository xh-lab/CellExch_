```r
import rpy2.robjects as ro
import rpy2.robjects.packages as rpackages

library(scImpute)
data_dir = "./"
out_dir = paste0(data_dir, "rerun/")
count_path = paste0(data_dir, "raw_count.csv")
dir.create(out_dir)
scimpute(count_path = count_path, infile = "csv", outfile = "csv",
         Kcluster = 9,
         out_dir = out_dir, drop_thre = 0.5, ncores = 36)
```

```r

count = read.csv(paste0(out_dir, "scimpute_count.csv"), row.names = 1)
saveRDS(count, file = paste0(out_dir, "zeisel_scimpute.rds"))
```
